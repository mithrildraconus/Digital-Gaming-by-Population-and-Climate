```python
import pandas as pd
import numpy as np

pd.set_option("max_columns", 200)

census_clean = pd.read_csv("Census Data (clean).csv")
climate_clean = pd.read_csv("Climate Data (clean).csv")

climate_census = pd.merge(census_clean, climate_clean, how='outer', on='Region')

climate_census = climate_census.drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis = 1)

climate_census.to_csv('Census-Climate Data (combo).csv')
```

```python
top100_clean = pd.read_csv("SteamSpy Top 100 Data (clean).csv")

num_groups = 9
top100_clean['Region'] = np.random.randint(num_groups, size=len(top100_clean))
top100_clean = top100_clean.drop(['Unnamed: 0'], axis = 1)

top100_region = top100_clean.replace({'Region': {0: 'Northeast',
                                                 1: 'Upper Midwest',
                                                 2: 'Ohio Valley',
                                                 3: 'Southeast',
                                                 4: 'Northern Rockies',
                                                 5:'South',
                                                 6:'Southwest',
                                                 7:'Northwest',
                                                 8:'West'}})

top100_region.to_csv('Top 100 Games & Regions Data.csv')
```

```python
region_merge = pd.merge(climate_census, top100_region, how='outer', on='Region')

print("Describe Data")
print(region_merge.describe())
print("\nSummarized Data")
print(region_merge.describe(include=['O']))
```

```python
cols_list = ['Region',
             'Location',
             'Total households',
             'Median household income (dollars) (in 2018 inflation-adjusted dollars)',
             'Mean household income (dollars) (in 2018 inflation-adjusted dollars)',
             'Total family households',
             'Median family income (dollars) (in 2018 inflation-adjusted dollars)',
             'Mean family income (dollars) (in 2018 inflation-adjusted dollars)',
             'Per capita income (dollars) (in 2018 inflation-adjusted dollars)',
             'Total nonfamily households',
             'Median nonfamily income (dollars) (in 2018 inflation-adjusted dollars)',
             'Mean nonfamily income (dollars) (in 2018 inflation-adjusted dollars)',
             'Median earnings for workers (dollars) (in 2018 inflation-adjusted dollars)',
             'Median earnings for male full-time, year-round workers (dollars) (in 2018 inflation-adjusted dollars)',
             'Median earnings for female full-time, year-round workers (dollars) (in 2018 inflation-adjusted dollars)',
             'Total population 16+ years',
             'Population 16+ years in labor force',
             'Population 16+ years in civilian labor force',
             'Month and Year',
             'Total precipitation (inches)',
             'Average temperature (°F)',
             'Game',
             'Total Steam owners (range)',
             'Mean Playtime since March 2009 (minutes)',
             'Mean Playtime for Last Two Weeks (minutes)',
             'Median Playtime since March 2009 (minutes)',
             'Median Playtime for Last Two Weeks (minutes)']

region_merge = region_merge.loc[:, cols_list]

print("Describe Data")
print(region_merge.describe())
print("\nSummarized Data")
print(region_merge.describe(include=['O']))
```

```python
region_merge.to_csv('Game-Census-Climate Data.csv')
```
